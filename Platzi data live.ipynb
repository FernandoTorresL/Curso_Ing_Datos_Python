### Extracci√≥n de informaci√≥n del HTML

```python
import bs4

soup = bs4.BeautifulSoup(response.text, 'html.parser')

print(soup.title.text)
```

```python
print(soup.select('meta[name=description]'))
```

```python
print(soup.select('meta[name=description]')[0]['content'])
```

```python
courses_links = soup.select('div[class=RecentCourses-content] > h3')

courses = [i.text for i in courses_links]

for i in courses:
    print('->', i)
```
